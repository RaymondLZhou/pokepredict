```python
import pandas as pd
import os

from PIL import Image
```

```python
df_enc = pd.read_csv("data\pokedex_enc.csv", index_col=0)

df_enc.info(verbose=1)
```

```python
cols = ['file'] + df_enc.columns[39:57].tolist()

cols
```

```python
df = pd.DataFrame(columns=cols, dtype='int64')

df.info()
```

```python
directory_1 = "data\images\POKEMON"

for filename in os.listdir(directory_1):
    file = os.path.splitext(filename)[0] + '.png'
    
    if os.path.splitext(filename)[-1] != '.png':
        print(filename)
        
        img = Image.open(os.path.join(directory_1, filename))
        img.save(os.path.join(directory_1, os.path.splitext(filename)[0]) + '.png')
        
        os.remove((os.path.join(directory_1, filename)))
        
    poke_num = int(os.path.splitext(filename)[0])
    row = df_enc[df_enc['pokedex_number'] == poke_num].iloc[0]
    
    entry = row[cols[1:]]
    entry['file'] = file
    df = df.append(entry)

df.head(n=10)
```

```python
directory_2 = "data\images\images"

for filename in os.listdir(directory_2):
    file = os.path.splitext(filename)[0] + '.png'
    
    if os.path.splitext(filename)[-1] != '.png':
        print(filename)
        
        img = Image.open(os.path.join(directory_2, filename))
        img.save(os.path.join(directory_2, os.path.splitext(filename)[0]) + '.png')
        
        os.remove((os.path.join(directory_2, filename)))
    
    name = os.path.splitext(filename)[0].capitalize()
    
    if "-" in name and (name.split("-")[1][0] != "o") and "ime" not in name and "Tapu" not in name or "Keldeo" in name: 
        if "Porygon" not in name and "Type" not in name:
            name = name.split("-")[0]
            
    mapping = {
        "Farfetchd" : "Farfetch'd",
        "Flabebe" : "Flabébé",
        "Mime-jr" : "Mime Jr.",
        "Mr-mime" : "Mr. Mime",
        "Porygon-z" : "Porygon-Z",
        "Tapu-bulu" : "Tapu Bulu",
        "Tapu-fini" : "Tapu Fini",
        "Tapu-lele" : "Tapu Lele",
        "Tapu-koko" : "Tapu Koko",
        "Type-null" : "Type: Null"
    }
    
    if name in mapping:
        name = mapping[name]
    
    row = df_enc[df_enc['name'].str.startswith(name)].iloc[0]
    
    entry = row[cols[1:]]
    entry['file'] = file
    df = df.append(entry)

df.tail(n=10)
```

```python
directory_3 = "data\images\PokemonData"

for directoryname in os.listdir(directory_3):  
    for filename in os.listdir(os.path.join(directory_3, directoryname)):      
        if os.path.splitext(filename)[-1] not in ['.png', '.jpg', '.jpeg']:
            os.remove(os.path.join(directory_3, directoryname, filename))
            print(directoryname, filename)
```

```python
directory_3 = "data\images\PokemonData"

for directoryname in os.listdir(directory_3):
    count = 0
    
    for filename in os.listdir(os.path.join(directory_3, directoryname)):      
        file = os.path.splitext(filename)[0]
        ext = os.path.splitext(filename)[-1]
        
        if filename != os.path.join(directory_3, directoryname, directoryname + '_' + str(count) + ext):
            os.rename(os.path.join(directory_3, directoryname, filename),
                     os.path.join(directory_3, directoryname, directoryname + '_' + str(count) + ext))
        
        count += 1
```

```python
directory_3 = "data\images\PokemonData"

for directoryname in os.listdir(directory_3):
    for filename in os.listdir(os.path.join(directory_3, directoryname)):      
        file = os.path.splitext(filename)[0] + '.png'
        
        if os.path.splitext(filename)[-1] != '.png':
            print(filename)

            img = Image.open(os.path.join(directory_3, directoryname, filename))
            img.save(os.path.join(directory_3, directoryname, os.path.splitext(filename)[0] + '.png'))

            os.remove((os.path.join(directory_3, directoryname, filename)))
        
        name = directoryname
        
        mapping = {
            "Farfetchd" : "Farfetch'd",
            "MrMime" : "Mr. Mime"
        }
        
        if name in mapping:
            name = mapping[name]
             
        row = df_enc[df_enc['name'] == name].iloc[0]

        entry = row[cols[1:]]
        entry['file'] = file
        df = df.append(entry)

df.tail(n=10)
```
